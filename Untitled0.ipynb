```python
%%bash
cat > bayes_classifier.py << 'PY'
#!/usr/bin/env python3
"""
Bayesian Decision Theory Classifier (Gaussian QDA / Gaussian Naive Bayes)
-------------------------------------------------------------------------

Usage (CLI):
    python bayes_classifier.py --csv data.csv --target target_col \
        --mode qda --standardize --plot --reg 1e-6

Required:
    --csv            Path to CSV file. All non-target columns are features.
    --target         Target column name or index (0-based).

Optional:
    --mode           "qda" (full covariance) or "gnb" (diagonal/naive). Default: qda
    --priors         Comma-separated priors in class order (e.g., "0.3,0.7"). If not set, empirical.
    --loss_csv       Path to CSV with LxL loss matrix (rows=true, cols=pred).
    --standardize    Standardize features by training mean/std.
    --reg            Diagonal covariance regularization (default 1e-6).
    --split          Test fraction (0..1). Default 0.3
    --seed           Random seed. Default 42
    --plot           Save a 2D decision plot using first two features.
    --out_prefix     Output prefix. Default "bdt_output"
"""

import argparse
import numpy as np
import pandas as pd
from typing import Tuple, Dict
import matplotlib.pyplot as plt

def _as_index(df: pd.DataFrame, target: str) -> int:
    try:
        idx = int(target)
        if idx < 0 or idx >= df.shape[1]:
            raise ValueError(f"Target index {idx} out of bounds.")
        return idx
    except ValueError:
        if target not in df.columns:
            raise ValueError(f"Target '{target}' not found. Available: {list(df.columns)}")
        return df.columns.get_loc(target)

def _encode_labels(y: pd.Series):
    uniques = pd.Index(sorted(y.unique(), key=lambda v: str(v)))
    mapping = {label: i for i, label in enumerate(uniques)}
    inv_mapping = {i: label for label, i in mapping.items()}
    y_enc = y.map(mapping).to_numpy()
    return y_enc, inv_mapping

def _compute_class_stats(X, y, mode="qda", reg=1e-6):
    classes = np.unique(y)
    d = X.shape[1]
    means, covs = [], []
    for k in classes:
        Xk = X[y == k]
        mu = Xk.mean(axis=0)
        Sk = np.cov(Xk, rowvar=False, bias=False)
        if np.ndim(Sk) == 0:
            Sk = np.array([[Sk]])
        if mode == "gnb":
            Sk = np.diag(np.diag(Sk))
        Sk = Sk + reg * np.eye(d)
        means.append(mu)
        covs.append(Sk)
    return np.array(means), np.array(covs)

def _log_gaussian_pdf(X, mu, Sigma):
    d = X.shape[1]
    try:
        inv = np.linalg.inv(Sigma)
        sign, logdet = np.linalg.slogdet(Sigma)
        if sign <= 0:
            raise np.linalg.LinAlgError
    except np.linalg.LinAlgError:
        jitter = 1e-3 * np.eye(d)
        inv = np.linalg.inv(Sigma + jitter)
        sign, logdet = np.linalg.slogdet(Sigma + jitter)
    diff = X - mu
    quad = np.einsum("...i,ij,...j->...", diff, inv, diff)
    return -0.5 * (d*np.log(2*np.pi) + logdet + quad)

def _posteriors_log(X, means, covs, priors):
    N, K = X.shape[0], means.shape[0]
    log_posts = np.zeros((N, K))
    for k in range(K):
        log_posts[:, k] = _log_gaussian_pdf(X, means[k], covs[k]) + np.log(priors[k])
    return log_posts

def _choose_min_risk(log_posts, loss):
    amax = np.max(log_posts, axis=1, keepdims=True)
    probs = np.exp(log_posts - amax)
    probs = probs / probs.sum(axis=1, keepdims=True)
    risks = probs @ loss
    return np.argmin(risks, axis=1)

def train_bdt(X, y, mode="qda", reg=1e-6, priors=None):
    y = np.asarray(y, dtype=int)
    means, covs = _compute_class_stats(X, y, mode=mode, reg=reg)
    K = len(np.unique(y))
    if priors is None:
        counts = np.bincount(y, minlength=K).astype(float)
        priors = counts / counts.sum()
    priors = np.asarray(priors, dtype=float)
    if not np.isclose(priors.sum(), 1.0):
        raise ValueError("Priors must sum to 1.")
    return {"means": means, "covs": covs, "priors": priors, "mode": mode}

def predict_bdt(model, X, loss=None):
    means, covs, priors = model["means"], model["covs"], model["priors"]
    K = means.shape[0]
    log_posts = _posteriors_log(X, means, covs, priors)
    if loss is None:
        return np.argmax(log_posts, axis=1)
    loss = np.asarray(loss, dtype=float)
    if loss.shape != (K, K):
        raise ValueError(f"Loss matrix must be {K}x{K}")
    return _choose_min_risk(log_posts, loss)

def accuracy(y_true, y_pred):
    return float((y_true == y_pred).mean())

def confusion_matrix(y_true, y_pred, K):
    cm = np.zeros((K, K), dtype=int)
    for t, p in zip(y_true, y_pred):
        cm[t, p] += 1
    return cm

def train_test_split(X, y, test_size=0.3, seed=42):
    rng = np.random.default_rng(seed)
    N = X.shape[0]
    idx = np.arange(N); rng.shuffle(idx)
    N_test = int(round(test_size * N))
    test_idx = idx[:N_test]; train_idx = idx[N_test:]
    return X[train_idx], X[test_idx], y[train_idx], y[test_idx]

def standardize_fit(X_train):
    mu = X_train.mean(axis=0)
    std = X_train.std(axis=0, ddof=0); std[std == 0] = 1.0
    return mu, std

def standardize_apply(X, mu, std):
    return (X - mu) / std

def main():
    import argparse, pandas as pd, numpy as np
    parser = argparse.ArgumentParser(description="BDT classifier (Gaussian QDA / Gaussian NB).")
    parser.add_argument("--csv", required=True)
    parser.add_argument("--target", required=True)
    parser.add_argument("--mode", choices=["qda","gnb"], default="qda")
    parser.add_argument("--priors", type=str, default=None)
    parser.add_argument("--loss_csv", type=str, default=None)
    parser.add_argument("--standardize", action="store_true")
    parser.add_argument("--reg", type=float, default=1e-6)
    parser.add_argument("--split", type=float, default=0.3)
    parser.add_argument("--seed", type=int, default=42)
    parser.add_argument("--plot", action="store_true")
    parser.add_argument("--out_prefix", type=str, default="bdt_output")
    args = parser.parse_args()

    df = pd.read_csv(args.csv)
    tgt_idx = _as_index(df, args.target)
    cols = list(df.columns)
    feature_cols = [c for i, c in enumerate(cols) if i != tgt_idx]
    X = df[feature_cols].to_numpy(dtype=float)
    y_raw = df.iloc[:, tgt_idx]

    y_enc, inv_map = _encode_labels(y_raw)
    K = len(inv_map)

    priors = None
    if args.priors is not None:
        vals = [float(v.strip()) for v in args.priors.split(",")]
        if len(vals) != K:
            raise ValueError(f"Priors length {len(vals)} != number of classes {K}")
        priors = np.array(vals, dtype=float)

    loss = None
    if args.loss_csv is not None:
        loss_df = pd.read_csv(args.loss_csv, header=None)
        loss = loss_df.to_numpy(dtype=float)
        if loss.shape != (K, K):
            raise ValueError(f"Loss matrix must be {K}x{K}, got {loss.shape}")

    X_train, X_test, y_train, y_test = train_test_split(X, y_enc, test_size=args.split, seed=args.seed)
    if args.standardize:
        mu, std = standardize_fit(X_train)
        X_train = standardize_apply(X_train, mu, std)
        X_test = standardize_apply(X_test, mu, std)

    model = train_bdt(X_train, y_train, mode=args.mode, reg=args.reg, priors=priors)
    y_pred = predict_bdt(model, X_test, loss=loss)

    acc = accuracy(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred, K)

    report = []
    report.append("Bayesian Decision Theory Classification Report\n")
    report.append(f"Mode: {args.mode.upper()}, Standardize: {args.standardize}, Reg: {args.reg}\n")
    report.append(f"Classes (encoded -> label): { {i: inv_map[i] for i in range(K)} }\n")
    report.append(f"Test Accuracy: {acc:.4f}\n")
    report.append("Confusion Matrix (rows=true, cols=pred):\n")
    report.append(str(cm))
    report_text = "\n".join(report)

    with open(f"{args.out_prefix}_report.txt", "w") as f:
        f.write(report_text)
    print(report_text)

    # Optional decision plot for first 2 features
    if args.plot and X.shape[1] >= 2:
        x_min, x_max = X_train[:, 0].min()-1.0, X_train[:, 0].max()+1.0
        y_min, y_max = X_train[:, 1].min()-1.0, X_train[:, 1].max()+1.0
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, 300), np.linspace(y_min, y_max, 300))
        grid = np.c_[xx.ravel(), yy.ravel()]
        d = X_train.shape[1]
        if d > 2:
            pad = np.zeros((grid.shape[0], d-2))
            pad[:] = X_train[:, 2:].mean(axis=0)
            grid_full = np.hstack([grid, pad])
        else:
            grid_full = grid
        preds = predict_bdt(model, grid_full, loss=loss).reshape(xx.shape)
        plt.figure()
        plt.contourf(xx, yy, preds, alpha=0.25)
        for k in range(K):
            pts = X_test[y_test == k]
            if pts.size > 0:
                plt.scatter(pts[:, 0], pts[:, 1], label=f"class {k} ({inv_map[k]})", s=18)
        plt.xlabel(f"{feature_cols[0]}"); plt.ylabel(f"{feature_cols[1]}")
        plt.legend(loc="best"); plt.title("BDT Decision Regions (first two features)")
        plt.savefig(f"{args.out_prefix}_plot.png", dpi=160, bbox_inches="tight")
        print(f"Saved plot: {args.out_prefix}_plot.png")

if __name__ == "__main__":
    main()
PY
```

```python
import numpy as np, pandas as pd

rng = np.random.default_rng(0)
N = 400
mu0 = np.array([0.0, 0.0, 0.0]); cov0 = np.array([[1.0, 0.6, 0.3],
                                                  [0.6, 1.2, 0.1],
                                                  [0.3, 0.1, 0.8]])
mu1 = np.array([2.0, 2.5, -0.5]); cov1 = np.array([[1.1, -0.4, 0.0],
                                                   [-0.4, 1.3, 0.2],
                                                   [0.0, 0.2, 0.9]])
X0 = rng.multivariate_normal(mu0, cov0, size=N//2)
X1 = rng.multivariate_normal(mu1, cov1, size=N//2)
X = np.vstack([X0, X1])
y = np.array(["A"]*(N//2) + ["B"]*(N//2))

df = pd.DataFrame(X, columns=["f1","f2","f3"])
df["label"] = y
df.to_csv("demo_bdt.csv", index=False)
df.head()
```
