```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score,
    RocCurveDisplay, PrecisionRecallDisplay
)
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from tensorflow.keras.datasets import mnist
import joblib

np.random.seed(42)
```

```python
# Load MNIST (already split into train and test by Keras)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Flatten 28x28 images → 784 features and scale to [0,1]
X_train = X_train.reshape(-1, 28*28).astype("float32") / 255.0
X_test  = X_test.reshape(-1, 28*28).astype("float32") / 255.0

# Turn into binary labels: 1 if target digit, else 0
TARGET_DIGIT = 5
y_train_bin = (y_train == TARGET_DIGIT).astype(int)
y_test_bin  = (y_test == TARGET_DIGIT).astype(int)

# Quick sanity check: class balance
print("Train positives (digit=={}):".format(TARGET_DIGIT), y_train_bin.sum(), "/", len(y_train_bin))
print("Test positives (digit=={}):".format(TARGET_DIGIT), y_test_bin.sum(), "/", len(y_test_bin))
```

```python
clf = Pipeline([
    ("scaler", StandardScaler(with_mean=False)),  # sparse-friendly; images are non-negative
    ("logreg", LogisticRegression(
        max_iter=1000,
        class_weight="balanced",   # helps with imbalance
        n_jobs=-1,
        solver="lbfgs"             # robust choice
    ))
])

clf.fit(X_train, y_train_bin)
```

```python
y_pred = clf.predict(X_test)

acc  = accuracy_score(y_test_bin, y_pred)
prec = precision_score(y_test_bin, y_pred, zero_division=0)
rec  = recall_score(y_test_bin, y_pred)
f1   = f1_score(y_test_bin, y_pred)

print(f"Accuracy:  {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1-score:  {f1:.4f}")

cm = confusion_matrix(y_test_bin, y_pred)
print("\nConfusion matrix (rows=true, cols=pred):\n", cm)

print("\nDetailed report:\n")
print(classification_report(y_test_bin, y_pred, target_names=["not-"+str(TARGET_DIGIT), str(TARGET_DIGIT)]))
```

```python
# For curves & AUCs you need a continuous score:
# Use decision_function if available, else predict_proba[:,1]
if hasattr(clf[-1], "decision_function"):
    scores = clf.decision_function(X_test)
else:
    scores = clf.predict_proba(X_test)[:, 1]

roc_auc = roc_auc_score(y_test_bin, scores)
print(f"ROC-AUC: {roc_auc:.4f}")

plt.figure()
RocCurveDisplay.from_predictions(y_test_bin, scores)
plt.title("ROC Curve (MNIST: {} vs rest)".format(TARGET_DIGIT))
plt.show()

plt.figure()
PrecisionRecallDisplay.from_predictions(y_test_bin, scores)
plt.title("Precision–Recall Curve (MNIST: {} vs rest)".format(TARGET_DIGIT))
plt.show()
```

```python
# Turn scores into probabilities if needed (LogReg decision_function is log-odds-ish; sigmoid is optional)
# But you can directly threshold on "scores". Lower threshold → more positives → higher recall, lower precision.
threshold = 0.0  # try -0.2, 0.2, etc. for decision_function; use 0.5 if you used predict_proba

y_pred_thresh = (scores >= threshold).astype(int)

print("With threshold =", threshold)
print("Precision:", precision_score(y_test_bin, y_pred_thresh, zero_division=0))
print("Recall:   ", recall_score(y_test_bin, y_pred_thresh))
print("F1:       ", f1_score(y_test_bin, y_pred_thresh))
print("CM:\n", confusion_matrix(y_test_bin, y_pred_thresh))
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(clf, X_train, y_train_bin, cv=cv, scoring="f1", n_jobs=-1)
print("5-fold CV F1 scores:", np.round(cv_scores, 4))
print("Mean F1:", cv_scores.mean().round(4))
```
